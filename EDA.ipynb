### Análisis 1: Accesos por Tecnología en Cada Provincia

En términos generales, se observa que la fibra óptica y el cablemódem son las tecnologías más utilizadas en la mayoría de las provincias, lo que indica un proceso de modernización en la infraestructura de telecomunicaciones. Existen marcadas diferencias entre provincias en cuanto al acceso a internet por tecnología. En Buenos Aires, Capital Federal, Córdoba y Santa Fe, se evidencia un predominio de las conexiones por fibra óptica y cablemódem, lo que sugiere una mayor densidad poblacional.

### Análisis 3: Comparación de Accesos a Internet por Provincia y Velocidad

```python
# Seleccionar las columnas de rangos de velocidad y la columna de provincia
columnas_rangos = [
    "HASTA 512 kbps", "+ 512 Kbps - 1 Mbps", "+ 1 Mbps - 6 Mbps",
    "+ 6 Mbps - 10 Mbps", "+ 10 Mbps - 20 Mbps", "+ 20 Mbps - 30 Mbps", "+ 30 Mbps", "OTROS"
]

# Obtener la lista de provincias
provincias = Accesos_por_rangos["Provincia"].unique()
num_provincias = len(provincias)

# Crear el gráfico de barras agrupadas
x = np.arange(num_provincias)  # Posiciones en el eje X
width = 0.1  # Ancho de las barras

plt.figure(figsize=(14, 7))

# Dibujar una barra para cada rango de velocidad
for i, rango in enumerate(columnas_rangos):
    plt.bar(x + i * width, Accesos_por_rangos.groupby("Provincia")[rango].sum(), width, label=rango)

plt.xlabel("Provincias")
plt.ylabel("Cantidad de Accesos")
plt.title("Comparación de Accesos a Internet por Provincia y Velocidad")
plt.xticks(x + width * (len(columnas_rangos) / 2), provincias, rotation=90)
plt.legend(title="Rangos de Velocidad")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

El gráfico de barras agrupadas muestra la distribución de accesos a internet en diferentes provincias, diferenciados por rango de velocidad, evidenciando disparidades en la infraestructura de telecomunicaciones. Las provincias con mayor cantidad de accesos se concentran en velocidades altas (30 Mbps o más), lo que sugiere una mayor disponibilidad de proveedores de internet con planes de alta velocidad, expansión de fibra óptica en zonas urbanas y mayor capacidad de gasto de los usuarios. En contraste, algunas provincias aún dependen de conexiones de menor velocidad (512 kbps o 1 Mbps - 6 Mbps), lo que puede reflejar falta de inversión en infraestructura o limitaciones económicas. En regiones rurales, el acceso a internet sigue estando restringido a tecnologías más antiguas como ADSL o redes móviles de baja velocidad, lo que acentúa la brecha digital y limita las oportunidades de desarrollo tecnológico en estas áreas.

### Análisis 5: Evolución de la Penetración de Internet en el Tiempo

```python

Penetración_poblacion["Periodo"] = Penetración_poblacion["Año"].astype(str) + " - T" + Penetración_poblacion["Trimestre"].astype(str)

# Agrupar los datos por periodo
Penetración_por_periodo = Penetración_poblacion.groupby("Periodo")["Accesos por cada 100 hab"].mean()

# Crear el gráfico de línea para mostrar la evolución de la penetración de internet
plt.figure(figsize=(12, 6))
plt.plot(Penetración_por_periodo.index, Penetración_por_periodo.values, marker="o", linestyle="-", color="b")

plt.xlabel("Periodo (Año - Trimestre)")
plt.ylabel("Accesos por cada 100 habitantes (Promedio)")
plt.title("Evolución de la Penetración de Internet en el Tiempo")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

El gráfico de barras refleja la cantidad de accesos a internet por cada 100 habitantes en distintas provincias, evidenciando diferencias significativas en la penetración del servicio. En la parte superior del gráfico se encuentran las provincias con mayor conectividad, como Capital Federal y Buenos Aires, donde la infraestructura tecnológica está más desarrollada. Estas regiones cuentan con una amplia oferta de proveedores de internet, lo que fomenta la competencia y mejora la calidad del servicio. Además, la presencia de redes de fibra óptica permite velocidades más altas, mientras que el mayor poder adquisitivo de los habitantes facilita la adopción de planes de internet premium. La alta concentración de empresas tecnológicas, oficinas y centros educativos en estas provincias también impulsa una mayor demanda de conectividad estable y de alta velocidad.

Por otro lado, las provincias que aparecen en la parte inferior del gráfico presentan una menor cantidad de accesos a internet por cada 100 habitantes, lo que sugiere una brecha digital significativa en el país. Factores como la falta de inversión en infraestructura, menores ingresos de la población y menor competencia entre proveedores pueden dificultar el acceso a una conexión de calidad en estas regiones. En muchos casos, los habitantes dependen de tecnologías más antiguas o de redes móviles con menor cobertura y estabilidad. Esta desigualdad en el acceso a internet limita las oportunidades de educación, trabajo y desarrollo digital, haciendo evidente la necesidad de políticas públicas e inversiones en telecomunicaciones para garantizar una conectividad más equitativa en todo el territorio nacional.


### Análisis 6: Evolución de Accesos a Internet por Tecnología

```python
# Verificar si la columna "Año" existe en el DataFrame antes de realizar modificaciones
if "Año" in Accesos_Por_Tecnología.columns:
    # Reemplazar NaN por una cadena vacía y eliminar cualquier carácter no numérico
    Accesos_Por_Tecnología["Año"] = (
        Accesos_Por_Tecnología["Año"]
        .astype(str)                         # Convertir a string por seguridad
        .str.replace(r'\D', '', regex=True)  # Eliminar cualquier carácter no numérico
    )

    # Eliminar filas donde "Año" quede vacío después de la limpieza
    Accesos_Por_Tecnología = Accesos_Por_Tecnología[Accesos_Por_Tecnología["Año"] != ""]

    # Convertir finalmente a entero
    Accesos_Por_Tecnología["Año"] = Accesos_Por_Tecnología["Año"].astype(int)

    # Agrupar por año y sumar los accesos por cada tecnología
    df_evolucion = Accesos_Por_Tecnología.groupby(["Año"])[["ADSL", "Cablemodem", "Fibra óptica", "Wireless", "Otros"]].sum()

    # Crear el gráfico de evolución de accesos a internet por tecnología
    plt.figure(figsize=(10, 6))
    df_evolucion.plot(kind="line", marker="o", colormap="tab10", ax=plt.gca())

    # Ajustes del gráfico
    plt.title("Evolución de Accesos a Internet por Tecnología")
    plt.xlabel("Año")
    plt.ylabel("Cantidad de Accesos")
    plt.xticks(rotation=45)
    plt.legend(title="Tecnología")
    plt.grid(True)
    plt.show()
else:
    print("La columna 'Año' no está disponible en el DataFrame.")

```

El gráfico de líneas muestra la evolución de los accesos a internet en Argentina a lo largo del tiempo, diferenciando las principales tecnologías utilizadas: ADSL, Cablemódem, Fibra Óptica, Wireless y Otros. Se observa un aumento sostenido en la adopción de fibra óptica, lo que refleja una tendencia hacia la modernización de la infraestructura de telecomunicaciones. Este crecimiento indica que las inversiones en redes de fibra están dando resultados y que los usuarios prefieren esta tecnología por su alta velocidad y estabilidad. En contraste, la cantidad de accesos por ADSL ha disminuido progresivamente, evidenciando un desplazamiento hacia tecnologías más avanzadas. Esto sugiere que la tecnología DSL está quedando obsoleta y siendo reemplazada gradualmente en muchas regiones.

Aunque el cablemódem no crece al mismo ritmo que la fibra óptica, sigue siendo una tecnología relevante, con una base estable de usuarios, especialmente en áreas donde representa la mejor opción para acceso a internet de alta velocidad. Por otro lado, las conexiones inalámbricas han mostrado fluctuaciones, lo que puede deberse a su uso en zonas rurales o en regiones donde la infraestructura de fibra óptica aún no está disponible, sirviendo como una alternativa temporal. Finalmente, la categoría "Otros" ha ido disminuyendo, lo que sugiere que tecnologías más antiguas o menos eficientes están siendo paulatinamente abandonadas a medida que los usuarios migran hacia soluciones más rápidas y confiables.

### KPI - Aumento del Acceso a Internet por Provincia (Proyección de +2%)

```python

# Asegurar que los nombres de columnas no contengan espacios extra
Penetración_poblacion.columns = Penetración_poblacion.columns.str.strip()

# Calcular el KPI de aumento del acceso a internet por provincia
Penetración_kpi = Penetración_poblacion.copy()

# Definir el aumento proyectado del 2%
Penetración_kpi["Nuevo Acceso"] = Penetración_kpi["Accesos por cada 100 hab"] * 1.02

# Calcular el KPI según la fórmula
Penetración_kpi["KPI (%)"] = ((Penetración_kpi["Nuevo Acceso"] - Penetración_kpi["Accesos por cada 100 hab"]) / 
                               Penetración_kpi["Accesos por cada 100 hab"]) * 100

# Agrupar por provincia y calcular el KPI promedio
Penetración_kpi_agrupado = Penetración_kpi.groupby("Provincia")["KPI (%)"].mean().reset_index()

# Graficar el KPI de aumento en acceso a Internet por provincia
plt.figure(figsize=(12, 6))
plt.barh(Penetración_kpi_agrupado["Provincia"], Penetración_kpi_agrupado["KPI (%)"], color="skyblue")

plt.xlabel("KPI (%) - Aumento del acceso a Internet")
plt.ylabel("Provincia")
plt.title("Aumento del Acceso a Internet por Provincia (Proyección de +2%)")
plt.gca().invert_yaxis()  # Invertir el eje para mostrar la provincia con mayor KPI arriba
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

### KPI - Crecimiento de Accesos por Tecnología

```python

# Definir la fórmula del KPI: Crecimiento de accesos por tecnología
# Crecimiento (%) = ((Accesos en trimestre actual - Accesos en trimestre anterior) / Accesos en trimestre anterior) * 100

# Ordenar los datos por año y trimestre
df_accesos_tecnologia = Totales_Accesos_Por_Tecnología.sort_values(by=["Año", "Trimestre"])

# Calcular el crecimiento porcentual por tecnología
df_accesos_tecnologia["Crecimiento ADSL"] = df_accesos_tecnologia["ADSL"].pct_change() * 100
df_accesos_tecnologia["Crecimiento Cablemodem"] = df_accesos_tecnologia["Cablemodem"].pct_change() * 100
df_accesos_tecnologia["Crecimiento Fibra Óptica"] = df_accesos_tecnologia["Fibra óptica"].pct_change() * 100
df_accesos_tecnologia["Crecimiento Wireless"] = df_accesos_tecnologia["Wireless"].pct_change() * 100

# Seleccionar los datos más recientes
latest_year = df_accesos_tecnologia["Año"].max()
latest_trimester = df_accesos_tecnologia[df_accesos_tecnologia["Año"] == latest_year]["Trimestre"].max()
df_latest = df_accesos_tecnologia[(df_accesos_tecnologia["Año"] == latest_year) & 
                                  (df_accesos_tecnologia["Trimestre"] == latest_trimester)]

# Filtrar solo las columnas de crecimiento
df_crecimiento = df_latest[["Crecimiento ADSL", "Crecimiento Cablemodem", "Crecimiento Fibra Óptica", "Crecimiento Wireless"]].T

df_crecimiento.columns = ["Crecimiento (%)"]

# Crear el gráfico de crecimiento por tecnología
plt.figure(figsize=(12, 6))
plt.bar(df_crecimiento.index, df_crecimiento["Crecimiento (%)"], color=['blue', 'green', 'red', 'orange'])
plt.xlabel("Tecnología")
plt.ylabel("Crecimiento (%)")
plt.title(f"Crecimiento de Accesos por Tecnología - {latest_year} T{latest_trimester}")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

El KPI de Crecimiento de Accesos por Tecnología permite evaluar la evolución del uso de diferentes tipos de conexión a Internet en el tiempo. Un crecimiento positivo en Fibra Óptica sugiere que más usuarios están optando por conexiones más rápidas y estables, lo que puede estar impulsado por mejoras en la infraestructura y mayores ofertas comerciales. Por otro lado, una caída en ADSL puede indicar una migración de usuarios hacia tecnologías más modernas, reflejando el declive de tecnologías más antiguas y menos eficientes. Cablemódem y Wireless pueden mostrar tendencias variadas, dependiendo de la disponibilidad y la demanda de conectividad en diferentes regiones.

El análisis del gráfico de crecimiento permite identificar qué tecnologías están en ascenso y cuáles están perdiendo mercado, lo que es clave para la toma de decisiones estratégicas. Si Wireless muestra un crecimiento, podría significar una mayor adopción en áreas rurales o regiones con menor infraestructura de fibra. Por otro lado, si Cablemódem se mantiene estable o en crecimiento, puede indicar que sigue siendo una alternativa viable en zonas urbanas. Estos datos pueden ayudar a las empresas a enfocar inversiones en tecnologías emergentes y a las autoridades a diseñar estrategias de conectividad más efectivas.

### KPI - Porcentaje de Usuarios con Conexión >100 Mbps

```python
# KPI: Porcentaje de Usuarios con Conexión de Alta Velocidad
# Cargar los datos de la hoja "Velocidad_sin_Rangos"
df_velocidad = Velocidad_sin_Rangos.sort_values(by=["Año", "Trimestre"])

# Seleccionar los datos más recientes
latest_year_vel = df_velocidad["Año"].max()
latest_trimester_vel = df_velocidad[df_velocidad["Año"] == latest_year_vel]["Trimestre"].max()
df_latest_vel = df_velocidad[(df_velocidad["Año"] == latest_year_vel) & 
                             (df_velocidad["Trimestre"] == latest_trimester_vel)]

# Calcular el porcentaje de usuarios con conexión >100 Mbps
usuarios_alta_velocidad = df_latest_vel[df_latest_vel["Velocidad"] > 100]["Accesos"].sum()
total_usuarios = df_latest_vel["Accesos"].sum()
porcentaje_alta_velocidad = (usuarios_alta_velocidad / total_usuarios) * 100

# Crear gráfico de porcentaje de usuarios con alta velocidad
plt.figure(figsize=(6, 6))
plt.pie([porcentaje_alta_velocidad, 100 - porcentaje_alta_velocidad], labels=[">100 Mbps", "<=100 Mbps"], 
        autopct="%.1f%%", colors=["blue", "gray"], startangle=90)
plt.title(f"Porcentaje de Usuarios con Conexión >100 Mbps - {latest_year_vel} T{latest_trimester_vel}")

# Mostrar el gráfico
plt.show()
```

El KPI de Porcentaje de Usuarios con Conexión de Alta Velocidad mide la proporción de usuarios con accesos superiores a 100 Mbps, reflejando el nivel de modernización de la infraestructura de Internet. Un alto porcentaje sugiere una creciente adopción de servicios de alta velocidad, lo que indica una mayor disponibilidad de fibra óptica o mejoras en otras tecnologías de conexión. En cambio, si el porcentaje es bajo, puede significar que gran parte de la población sigue dependiendo de conexiones más lentas, posiblemente debido a barreras económicas o limitaciones en la cobertura de redes avanzadas.

Este indicador es clave para evaluar la equidad en el acceso a Internet y planificar estrategias de expansión de infraestructura. Si la adopción de alta velocidad es baja, sería recomendable incentivar inversiones en redes de fibra óptica y ofrecer planes más accesibles para los usuarios. Además, su seguimiento en el tiempo permitirá analizar si las políticas de conectividad están logrando mejorar la calidad del servicio y reducir la brecha digital en distintas regiones.
